```python
# @TODO: Import libraries
from pathlib import Path
import csv

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('menu_data.csv')
sales_filepath = Path('sales_data.csv')

# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []

# @TODO: Read in the menu data into the menu list
with open(menu_filepath, 'r') as menu_file:
    csv_reader_menu = csv.reader(menu_file, delimiter=",")
    header_menu = next(csv_reader_menu)
    menu.append(header_menu)

    for row in csv_reader_menu:
         menu.append(row)

# @TODO: Read in the sales data into the sales list
with open(sales_filepath, 'r') as sales_file:
    csv_reader_sales = csv.reader(sales_file, delimiter=",")
    header_sales = next(csv_reader_sales)
    sales.append(header_sales)

    for row_sales in csv_reader_sales:
        sales.append(row_sales)


# @TODO: Initialize dict object to hold our key-value pairs of items and metrics

report = {}

# Initialize variables
quantity_s = 0
quantity = 0
price = 0
cost = 0
count = 0
revenue = 0
cogs = 0
profit = 0

# Since the sales list gives data of sales item and quantity while menu list does items and each price / cost, we need to make a nested 
#  directory first with its key as sales_item and sum of quantity for each items. 
  
# Loop over every row in the sales list object, using a conditional if the item value not in the report, 
# add it as a new entry with initialized metrics
# Naming convention allows the keys to be ordered in logical fashion, count, revenue, cost, profit
for s in sales[1:]:
           quantity_s = int(s[3])
           sales_item = s[4]
           if sales_item not in report.keys():
               report[sales_item] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0}
                       
           report[sales_item]["01-count"] += quantity_s 
  
# For every row in our sales data, loop over the menu records to determine a match
    
for m in menu[1:]:
           item = m[0]
        
#     If the item value in our sales data is equal to the any of the items in the menu, then begin tracking metrics for that item       
          
           for sales_item in report:
              if item == sales_item:
                 price = float(m[3])
                 cost = float(m[4])
                 profit = price - cost  # Calculate profit of each item in the menu data
                 quantity = report[sales_item]["01-count"] # returns total quantity value for further calculation

# Cumulatively add up the metrics for each item key
                 report[sales_item]["02-revenue"] += price * quantity
                 report[sales_item]["03-cogs"] += cost * quantity
                 report[sales_item]["04-profit"] += profit * quantity 

# Else, the sales item does not equal any fo the item in the menu data, therefore no match               
              else:
                 continue
print(report)  # check the result

# Write out report to a text file (won't appear on the command line output)
output_path = Path('output.txt')

with open(output_path, 'w') as file:
    csv_writer = csv.writer(file)
    for x in report.items():
        csv_writer.writerow(x)
 
```
