```python
# import the pathlib and csv library
from pathlib import Path
import csv
# Set the file path 
filepath = Path("budget_data.csv")

# Initialize list of records
records = []

# Initialize variables
total_month = 0
total_pnl = 0
total_change = 0
count_change = 0
minimum = 0
minimum_index = 0
maximum = 0
maximum_index = 0

# Initialize dictionary to get the index information to calculate the changes in profit and loss
change_dict = {} 
    
# Open the csv file as an object
with open(filepath, 'r') as file:

# Pass in the csv file to the csv.reader() function and return the csvreader object
      csvreader = csv.reader(file, delimiter=",")

# Read the header row and append 'Change in Profit and loss' for containing new values.
      header = next(csvreader)
      header.append('Change in Profit and Loss')
    
# Append the header to the list of records
      records.append(header)
   
# Read each row of data after the header. 
# However, we need to calculate between values in different lines. To to the job during iteration, 
# we need to use enumerate() to get the index information so that we can call values in different lines.

      for index, row in enumerate(csvreader):
           date = row[0] # The type of 'row' is list with two elements at this point. date is at index 0.
           profit_loss = int(row[1])  # need to convert the profit_loss data into integer for calculation
           total_month += 1  # calculate total number of months
           total_pnl += profit_loss  # calculate total profit/loss by summing up all values
           change_dict[index] = int(row[1]) # This dictionary will add index as keys and profit_loss data as values
 
# we need conditional since the first date will not generate an appropriate change in profit/loss(no previous day data)
           if   index == 0:
                row.append("N/A")  # so we need to put "N/A" for change at the first date
                records.append(row) 
               
           elif index >= 1:      # From the second date, we can calculate the changes in daily profit/loss
                change = change_dict[index] - change_dict[index - 1]
                total_change += change  # total change is necessary for calculation of average change
                count_change += 1 # also necessary for calculation of average change(first date must not be regarded)
                row.append(change)
                records.append(row)
                
# We need another conditional to find the greatest increase/decrease date and value 
                if change <= minimum:
                     minimum = change
                     minimum_index = index + 1   # "index + 1" for taking into account of the first date, which is valued "N/A"
                elif change > maximum:
                     maximum = change
                     maximum_index = index + 1

# calculate the average changes
average = round(total_change / count_change, 2)
# find the greatest increase/decrease date by index in the nested list of records.
great_increase_date = records[maximum_index][0]
great_decrease_date = records[minimum_index][0]

# Set the output file path for output.txt.
output_path = Path("output.txt")

# Write the contents of the list of records to output.txt using csv.writer(). 
with open(output_path, 'w') as file:
    csv_writer = csv.writer(file, delimiter=",")
    for x in records:
        csv_writer.writerow(x)
        
# print the result of,
# The total number of months included in the dataset.
# The net total amount of Profit/Losses over the entire period.
# The average of the changes in Profit/Losses over the entire period.
# The greatest increase in profits (date and amount) over the entire period.
# The greatest decrease in losses (date and amount) over the entire period.

print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {total_month}")
print(f"Total: ${total_pnl}")
print(f"Average Change: ${average}")
print(f"Greatest Increase in Profits: {great_increase_date} (${maximum})")
print(f"Greatest Decrease in Profits: {great_decrease_date} (${minimum})")
```
